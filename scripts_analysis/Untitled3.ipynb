```python
import csv 
import pandas as pd 

best = pd.read_csv("analysis_baseline_ours_w_queris_best.csv", header=None)
before = pd.read_csv("analysis_baseline_ours_w_queris.csv", header=None)
```

```python
befores = {} 
for _,row in before.iterrows():
    if pd.isnull(row[1]):
        continue 
    _id = row[1]
    befores[_] = [row[2], row[3]]
```

```python
bests = {} 
for _,row in best.iterrows():
    if pd.isnull(row[1]):
        continue 
    _id = row[1]
    bests[_] = [row[2], row[3]]
    
```

```python
improv = 0
big_improv = 0 
lower = 0
improv_list = [] 
big_improv_list = [] 
lower_list = [] 
same = [] 
for k,v in bests.items():
    new_v = v[0]
    base_v = befores[k][0]
    if new_v == "R7" and (base_v == "R3" or base_v == "Else"):
        big_improv_list.append(k)
    elif new_v == "R7" and base_v == "R5":
        improv_list.append(k)
    elif base_v == "R7" and (new_v == "R3" or new_v == "Else" or new_v == "R5"):
        lower_list.append(k)
    else:
        if new_v == base_v:
            same.append(k)
        
```
